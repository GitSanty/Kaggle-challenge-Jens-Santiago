```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

df=pd.read_csv('Netflix Shows.csv', encoding='cp437')
df.dropna(inplace=True)


plt.figure(figsize=(8,8))

x=['humor','scary','sex','action','fantasy']
y=[df[df.ratingLevel.str.contains('humor')]['user rating score'].mean(),
   df[df.ratingLevel.str.contains('scary')]['user rating score'].mean(),
   df[df.ratingLevel.str.contains('sex')]['user rating score'].mean(),
   df[df.ratingLevel.str.contains('action')]['user rating score'].mean(),
   df[df.ratingLevel.str.contains('fantasy')]['user rating score'].mean() ]
plt.plot(x, y)

'''y2=[df[df.ratingLevel.str.contains('humor')]['user rating score'].to_list(),
   df[df.ratingLevel.str.contains('scary')]['user rating score'].to_list(),
   df[df.ratingLevel.str.contains('sex')]['user rating score'].to_list(),
   df[df.ratingLevel.str.contains('action')]['user rating score'].to_list(),
   df[df.ratingLevel.str.contains('fantasy')]['user rating score'].to_list() ]'''

df.loc[df.ratingLevel.str.contains('humor'), 'genre']='humor'
df.loc[df.ratingLevel.str.contains('scary'), 'genre']='scary'
df.loc[df.ratingLevel.str.contains('sex'), 'genre']='sex'
df.loc[df.ratingLevel.str.contains('action'), 'genre']='action'
df.loc[df.ratingLevel.str.contains('fantasy'), 'genre']='fantasy'
df.dropna()
sns.swarmplot(x=df.genre, y=df['user rating score'])
plt.ylabel('user rating score', fontdict={'fontsize':20})
plt.xlabel('user rating score', fontdict={'fontsize':20})

plt.title('average user rating by genre', fontdict={'fontsize':25})
plt.show()
df.shape()
```
